# 1. Project Problem and Hypothesis

This project seeks to predict the impact of a restaurant's health inspection scores and online profile quality on the number of Foursquare checkins it receives. This analysis will help restaurants, esp. family owned restaurants, to focus their marketing and quality efforts on the the most impactful areas that truly brings customers in. 

The feature matrix will include the following fields:


| Variable |    Description|  Type |
|----------|:-------------:|------:|
| inspection_score |  average inspection score since 2014 | continuous |
| price_tier |    number between 1 and 4  where 4 = Most Expensive  |   discrete |
| has_menu | 0 = no, 1 = yes indicating whether menu is available on Foursquare|   boolean|
| rating | floats between 0 and 10 where 10 is the highest average rate |   continuous|
| number_ratings | number of ratings |   continuous|
| number_photos | number of photos on their Foursquare profile |   continuous|
| reservations | 0 = no, 1 = yes. Indicates if restaurant takes reservations |   boolean|
| credit_cards | 0 = no, 1 = yes. Indicates if restaurant takes credit cards |   boolean|
| number_reviews | count of "tips" or reviews posted on Foursquare |   continuous|
| outdoor_seating | 0 = no, 1 = yes. Indicated whether outdoor seating is available |   boolean|
| years_on_foursquare | number of years on foursquare |   continuous|

The response vector will include the following field:

| Variable |    Description|  Type |
|----------|:-------------:|------:|
| checkin_count |  total number of checkins to date | continuous |

## Foursquare Data

The second set of data is pulled from the Foursquare Search API. In order to pull the restaurant level data, I will need to use these 2 endpoints:
https://developer.foursquare.com/docs/venues/search
https://developer.foursquare.com/docs/responses/venue

The search endpoint is to map the list of restaurant names from the DataSF database to the list of restaurant names on Foursquare. Given that the Foursquare search endpoint does not require an exact match, I do not expect to find too many null values. This however is to be seen. The output of this first query will be a list containing 
* foursquare_IDs
* foursquare_name
* dataSf_Id
* dataSf_name

## 4. Project Concerns

I have a few concerns about this project
* There may be no signal in the data and this whole project is futile
* The factors I am considering are irrelevant and I am missing key features (such as the content of the reviews itself)
* Users who checkin at Foursquare represent a non-random sample of the population. While this model may predict foursquare checkins, it may not represent overall foot traffic
* Foursquare checkins dont tell us the group size. A group of 10 vs a group of 3 may have 1 check in each
* If the model is incorrect and used, restaurant owners may invest in beefing up their online profiles or health inspection scores without much result in foot traffic. However, there isn't really a big downside outside of any costs incurred due to this
